```python
def sql_to_pandas(sql_query, database = connection):
    table = pd.read_sql_query(sql_query, database)
    return table
```

```python
loan_county_pdf = sql_to_pandas("""
                                SELECT 
                                            funded_at as funded_dt
                                            , loan_amount
                                            , rate
                                            , l.state
                                            , l.county_fips as fips
                                            , loan_limit
                                FROM 
                                            loans l
                                LEFT JOIN
                                            counties c
                                on
                                            l.state = c.state AND
                                            l.county_fips = c.fips
                                order by
                                            funded_dt, state
                          """)
```

```python
loan_county_pdf['better_loan_size'] = np.where(loan_county_pdf['loan_limit'] > loan_county_pdf['loan_amount']
                                               , 'conforming'
                                               , 'jumbo')
```

# How has the loan interest rate varied over time?

```python
loan_county_pdf['rate'].plot.kde()
plt.show()
```

```python
loan_county_pdf['simple_interest_yearly'] = .01*loan_county_pdf['loan_amount']*loan_county_pdf['rate']
```

```python
daily_rate_avg = loan_county_pdf.groupby('funded_dt')['loan_amount','simple_interest_yearly'].apply(lambda x: x.sum()).reset_index()
```

```python
daily_rate_avg['simple_interest_rate_daily_avg'] = 100*daily_rate_avg['simple_interest_yearly'] / daily_rate_avg['loan_amount']
```

```python
#get rid of superfluous entries
daily_rate_avg = daily_rate_avg[3:10000]
```

```python
#set date as index
daily_rate_avg.set_index('funded_dt', inplace=True)
```

```python
# add 7-day rolling for smoothness
daily_rate_avg['loan_amount_7d_rolling'] = daily_rate_avg['loan_amount'].rolling(window='7D').sum()
daily_rate_avg['simple_interest_yearly_7d_rolling'] = daily_rate_avg['simple_interest_yearly'].rolling(window='7D').sum()
```

```python
#compute 7 day rolling simple interest rate as 7-day rolling yearly simple interest over 7 days rolling loan amount
daily_rate_avg['simple_interest_rate_7d_rolling'] = 100*daily_rate_avg['simple_interest_yearly_7d_rolling']/daily_rate_avg['loan_amount_7d_rolling']
```

```python
def conf_low(interest_df):
    conf_int = proportion_confint(interest_df['simple_interest_yearly_7d_rolling'],interest_df['loan_amount_7d_rolling'],.05, method='normal')
    return conf_int[0]

def conf_high(interest_df):
    conf_int = proportion_confint(interest_df['simple_interest_yearly_7d_rolling'],interest_df['loan_amount_7d_rolling'],.05, method='normal')
    return conf_int[1]
```

```python
daily_rate_avg['conf_low'] = 100*daily_rate_avg.apply(conf_low, axis=1)
daily_rate_avg['conf_high'] = 100*daily_rate_avg.apply(conf_high, axis=1)
```

# 7 Day Rolling Simple Interest Rate

```python
plt.rcParams.update({'font.size': 22})
fig = plt.figure(figsize=(30,20))
plt.plot(daily_rate_avg['simple_interest_rate_7d_rolling'], color='black')
plt.fill_between(daily_rate_avg.index, daily_rate_avg['conf_low'], daily_rate_avg['conf_high'])
plt.title("7 Day Rolling Simple Interest Rate")
plt.xticks(rotation='vertical')
plt.show()
```

# Takeaways
* Interest rates
    * Increased from mid 2016 to mid 2017;
    * Stabilized through the end of 2017;
    * Increased again through 2018;
    * Then dropped back to second half of 2017 levels, briefly stabilizing toward the end of 2019
* The blue shading gives the 95% confidence interval based on considering each dollar of the loan a Bernoulli trial, and each simple interest rate point a success. With this interpretation we see our rolling simple interest rate measurement has very little uncertainty due to the high dollar amount of the loans.
* However, since rolling simple interest rate is in fact a ratio of sums, we don't have a straightforward method for rigorously computing uncertainty. We would expect it to be somewhat higher due to the low number of observations in the measurement.

# How much does the interest rate differ between conforming and jumbo?

```python
loan_size_rate_avg = loan_county_pdf.groupby('better_loan_size')['loan_amount','simple_interest_yearly'].apply(lambda x: x.sum()).reset_index()
```

```python
loan_size_rate_avg['loan_size_simple_interest_rate_avg'] = 100*loan_size_rate_avg['simple_interest_yearly']/loan_size_rate_avg['loan_amount']
```

```python
def conf_low_loan_size(interest_df):
    conf_int = proportion_confint(interest_df['simple_interest_yearly'],interest_df['loan_amount'],.05, method='normal')
    return conf_int[0]

def conf_high_loan_size(interest_df):
    conf_int = proportion_confint(interest_df['simple_interest_yearly'],interest_df['loan_amount'],.05, method='normal')
    return conf_int[1]
```

```python
loan_size_rate_avg['conf_low'] = 100*loan_size_rate_avg.apply(conf_low_loan_size, axis=1)
loan_size_rate_avg['conf_high'] = 100*loan_size_rate_avg.apply(conf_high_loan_size, axis=1)
```

```python
loan_size_rate_avg['error_top'] = loan_size_rate_avg['conf_high'] - loan_size_rate_avg['loan_size_simple_interest_rate_avg']
loan_size_rate_avg['error_bottom'] = loan_size_rate_avg['loan_size_simple_interest_rate_avg'] - loan_size_rate_avg['conf_low'] 
```

```python
loan_size_rate_avg
```

```python
fig = plt.figure(figsize=(30,20))
plt.bar(loan_size_rate_avg['better_loan_size']
        , loan_size_rate_avg['loan_size_simple_interest_rate_avg']
        , yerr=(loan_size_rate_avg['error_top'], loan_size_rate_avg['error_bottom'])
        , color=['blue','orange'])
plt.title("Average Simple Interest Rate by Loan Size")
plt.show()
```

# Takeaways
* Roughly 96.7% of loans are conforming, 3.3% jumbo
* Conforming loans have, on average, a 4.5% higher simple interest rate than jumbo loans
* If we compute the error based on Bernoulli trials as before, we find this result statistically significant at 95% confidence. The error bars are so small you can barely see them.
* However, as before, we cannot interpretably compute a rigorous error bound for simple interest rate. Since this average contains more observations, we expect less uncertainty, but we can't say for sure the difference is signicant because it is relatively small.
