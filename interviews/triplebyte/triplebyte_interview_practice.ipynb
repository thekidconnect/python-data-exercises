```python
class Leaf:
    def __init__(self, label):
        self.label = label

class Node:
    def __init__(self, column, split_value, left_subtree, right_subtree):
        self.column = column
        self.split_value = split_value
        self.left_subtree = left_subtree
        self.right_subtree = right_subtree
```

```python
def predict(node, x):
    if isinstance(node, Leaf):
        print("Is Leaf")
        return node.label
    if x[node.column] < node.split_value:
        return predict(node.left_subtree, x)
    else:
        return predict(node.right_subtree, x)
```

```python
tree = Node(0, 5.6, Leaf(0),Node(2, 5.0, Leaf(1),Leaf(2)))
```
